#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowel(line):
    return sum(1 for symbol in line.lower() if symbol in 'aeiou')

def run_tests():
    assert count_vowel('aeiou') == 5, "Ошибка в тесте 1"
    assert count_vowel(' ') == 0, "Ошибка в тесте 2"
    assert count_vowel('AEIOU') == 5, "Ошибка в тесте 3"
    print("Все тесты пройдены успешно")

run_tests()
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def has_unique_chars(text):
    return len(text) == len(set(text))

def run_tests():
    assert has_unique_chars('abcde') == True, "Ошибка в тесте 1"
    assert has_unique_chars('aaaaaa') == False, "Ошибка в тесте 2"
    assert has_unique_chars('hello') == False, "Ошибка в тесте 3"
    print("Все тесты пройдены успешно")

run_tests()
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_ones(num):
    if num <= 0:
        raise ValueError("Число должно быть положительным")
    return bin(num).count('1')

def run_tests():
    assert count_ones(1) == 1, "Ошибка в тесте 1"
    assert count_ones(12) == 2, "Ошибка в тесте 2"
    assert count_ones(63) == 6, "Ошибка в тесте 3"
    print("Все тесты пройдены успешно")

run_tests()
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    if n <= 0:
        raise ValueError("Число должно быть положительным")
    
    result = 1
    iterations = 0
    while n >= 10:
        result = 1
        for digit in str(n):
            result *= int(digit)
        n = result
        iterations += 1
    
    return iterations

def run_tests():
    assert magic(39) == 3, "Ошибка в тесте 1"
    assert magic(999) == 4, "Ошибка в тесте 2"
    assert magic(2) == 0, "Ошибка в тесте 3"
    print("Все тесты пройдены успешно")

run_tests()
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, targets):
    if len(pred) != len(targets):
        raise ValueError("Векторы должны быть одинаковой длины")
    
    sum_squared_diff = sum((p - t) ** 2 for p, t in zip(pred, targets))
    return (sum_squared_diff / len(pred)) ** 0.5

def run_tests():
    assert mse([1, 2, 3], [1, 2, 3]) == 0, "Ошибка в тесте 1"
    assert mse([5], [3]) == 2, "Ошибка в тесте 2"
    assert mse([10, 20, 30], [5, 15, 25]) == 5, "Ошибка в тесте 3"
    print("Все тесты пройдены успешно")

run_tests()
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def factorize(n):
    if n <= 0:
        raise ValueError("Число должно быть положительным")
    if n == 1:
        return "(1)"
    
    factors_list = []
    divisor = 2
    
    while divisor * divisor <= n:
        power = 0
        while n % divisor == 0:
            power += 1
            n //= divisor
        if power:
            factors_list.append((divisor, power))
        divisor += 1
    
    if n > 1:
        factors_list.append((n, 1))
    
    return "".join(f"({f})" if p == 1 else f"({f}**{p})" for f, p in factors_list)

def run_tests():
    assert factorize(123) == "(3)(41)", "Ошибка в тесте 1"
    assert factorize(86240) == "(2**5)(5)(7**2)(11)", "Ошибка в тесте 2"
    assert factorize(1) == "(1)", "Ошибка в тесте 3"
    print("Все тесты пройдены успешно")

run_tests()
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(n):
    if n <= 0:
        return "It is impossible"
    
    level = 1
    total = 0
    
    while total < n:
        total += level ** 2
        if total == n:
            return level
        level += 1
    
    return "It is impossible"

def run_tests():
    assert pyramid(1) == 1, "Ошибка в тесте 1"
    assert pyramid(2) == "It is impossible", "Ошибка в тесте 2"
    assert pyramid(55) == 5, "Ошибка в тесте 3"
    print("Все тесты пройдены успешно")

run_tests()
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(num):
    if num <= 0:
        raise ValueError("Число должно быть положительным")
    
    digits_str = str(num)
    size = len(digits_str)
    
    if size % 2 == 0:
        left_end = size // 2
        right_start = left_end
    else:
        left_end = size // 2
        right_start = size // 2 + 1
    
    left_sum = sum(int(d) for d in digits_str[:left_end])
    right_sum = sum(int(d) for d in digits_str[right_start:])
    
    return left_sum == right_sum

def run_tests():
    assert is_balanced(121) == True, "Ошибка в тесте 1"
    assert is_balanced(66575) == True, "Ошибка в тесте 2"
    assert is_balanced(11345) == False, "Ошибка в тесте 3"
    print("Все тесты пройдены успешно")

run_tests()
```
